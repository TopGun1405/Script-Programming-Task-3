```
    s = '  first star   :   second star   :    third star  '
    l = ['first star', 'second star', 'third star']
```
    - 1) for ~ in 문을 이용한 코드를 작성하시오.
    - 2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

- 1)

````python
s = '  first star   :   second star   :    third star  '
l = []

index = 0
for i in range(len(s)) :
    if s[i] == ':' or i == len(s) - 1:
        st = ''.join(s[index : i]).strip()
        l.append(st)
        index = i + 1

l
````

- ':' 문자와 맨 마지막 문자의 인덱스 값을 통해 문자 양옆의 공백을 제거해 리스트에 추가하는 것을 for ~ in으로 구현하였습니다.

- 2)

````python
s = '  first star   :   second star   :    third star  '

index = [i for i in range(-1, len(s)) 
         if i == -1 or s[i] == ':' or i == len(s) - 1]

l = [s[index[i] + 1 : index[i + 1]].strip() for i in range(len(index)) 
     if i < len(index) - 1]

l
````

- 리스트 내포를 이용해 1)에서 구현한 기능을 그대로 구현하였습니다.

5. 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.
    - 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.)
        -  ```
          >>> addall([1])
          1
```
        - ```
          >>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
          45
```
    - 2) addallodd(list) 함수 (리스트 내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
        - ```
          >>> addallodd([1])

        - ```
          >>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])

```
    - [참고 1]: 리스트 내포 방식으로 addall() 및 addallodd() 함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)
    - [참고 2]: sum() 등의 내장함수 사용 가능함

- 1)

````python
def addall(list) :
    sum_list = []
    for i in list :
        sum_list.append(i)
    return sum(sum_list)

print(addall([1]))
print(addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
````

````python
def addall(list) :
    return sum([i for i in list])

print(addall([1]))
print(addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
````

- 2)

````python
def addallodd(list) :
    sumodd_list = []
    for i in list :
        if i % 2 == 1 :
            sumodd_list.append(i)
    return sum(sumodd_list)

print(addallodd([1]))
print(addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
````

````python
def addallodd(list) :
    return sum([i for i in list if i % 2 == 1])

print(addallodd([1]))
print(addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
````

8. 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.
    - length(X)를 다음과 같이 정의한다 : 10진수로 표현된 X의 자릿수
    - Strange라는 숫자를 아래와 같이 정의한다.
        - 음이 아닌 정수 중 자릿수가 1인 모든 숫자는 Strange이다.
        - length(X) > 1 인 숫자 X는 다음 2가지를 만족하면 Strange이다.
            - X는 length(X)로 정확히 나누어 떨어진다.
            - X / length(X) 가 Strange이다.
            
    - 이 문제에서 숫자 하나가 주어질 때, 이 숫자가 Strange인지 판단하는 함수 isStrange(num)을 작성하시오.
    - isStrange(num) 함수가 재귀적인 함수가 되도록 작성해 보세요.

````python
def isStrange(num) :
    if length(num) > 1 and num % length(num) == 0 :
        if length(num / length(num)) != 1 :
            isStrange(num / length(num))
        return True
    if length(num) == 1 and num % length(num) == 0:
        return True
    else :
        return False

def length(X) :
    return len(str(X))

print(isStrange(0))
print(isStrange(56))
print(isStrange(100))
print(isStrange(7882602506492313600))
print(isStrange(6012302549522098651))
````

9. 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.

    - 1번 그룹, 2번 그룹, ... n번 그룹으로 번호가 매겨진 총 n개의 그룹이 존재하고 $i^{th}$그룹에는 임의의 팀원 수 $a_i$ 명이 있다. 하나의 버스 정류장이 있고, 버스는 오직 한 대만 운영한다. 버스는 버스 정류장에 도착하고 모든 그룹은 버스를 통해 여행하길 원한다.<br>
    하지만, 같은 그룹에 속한 팀원들은 분리되기 원치 않는다. 그래서 그룹의 모든 팀원이 탈 수 있을 때 에만 그 그룹은 버스에 탑승한다.  
    또한, 이들은 그룹의 번호순으로 여행을 하길 원한다. 예를 들어, 3번 그룹은 1번 그룹과 2번 그룹이 여행을 한 뒤에 버스를 탈 수 있거나, 1번 그룹과 2번 그룹이 이미 버스에 탑승한 상태에 있을 때 동시에 버스를 탈 수 있다.
    마지막으로, 버스는 빈 좌석이 없을 때 에만 여행을 떠난다.<br><br>
    
    - 이 문제에서 각 그룹의 팀원 수가 리스트로 주어질 때, 가능한 모든 버스의 사이즈 x(x명의 사람들이 동시에 탈 수 있음)을 반환하는 함수 solve(a) 를 작성하시오.
    - [주의] 반드시 계산 복잡도를 최대한 낮추세요. 아래 예시 3번을 실행했을 때 결과가 바로 나타날 수 있도록!!

````python
def solve(a) :
    bus_size = []
    total_p = sum(a)
    max_group_p = max(a)
    
    near_group = 0
    for i in a :
        near_group += i
        if total_p % near_group == 0 and near_group >= max_group_p :
            bus_size.append(near_group)
    return bus_size
    
print(solve([1]))
print(solve([1, 2, 1, 1, 1, 2, 1, 3]))
print(solve([1, 1, 1, 1]))
print(solve([1, 3, 4, 1, 1, 2]))
print(solve([1, 5, 2, 3, 1, 2, 5]))
print(solve([2, 7, 23, 2, 1, 7, 5, 35]))
print(solve([9999] * 10 ** 5))
````

````python
import requests
import string

req = requests.get('http://link.koreatech.ac.kr/')
source = req.text
punct = string.punctuation

# 추후 숙제 답안 제출 때엔 아래 코드는 주석처리할 것
# print(source)
# print(punct)

def return_word_Num_dictionary(link) :
    req = requests.get(link)
    source = req.text
    
    htmlTag = 0
    pureStr = ''
    pureStr_list = []
    css = ''
    jvsc = ''
    for i in range(len(source)) :
        if source[i] == '<' and htmlTag == 0 :
            # 요구사항 1
            for j in range(1, 6) :
                css += source[i]
            for k in range(1, 7) :
                jvsc += source[i]
            if css != 'style' or jvsc != 'script' :
                htmlTag = 1
        if source[i] == '>' and htmlTag == 1 :
            htmlTag = 0
            continue
        if htmlTag == 0 :
            pureStr += source[i]
            
    pureStr_list = pureStr.split()
    
    # 요구사항 2
    for i in range(len(pureStr_list)) :
        pureStr_list[i] = pureStr_list[i].strip().strip(punct)
        
    for i in range(len(pureStr_list)) :
        for j in punct :
            while j in pureStr_list[i] :
                for k in range(len(pureStr_list[i])) :
                    indx = pureStr_list[i].find(j)
                    pureStr_list[i] = pureStr_list[i][ : indx] + \
                    pureStr_list[i][indx + 1 : ]
                    break
                    
    # 요구사항 3
    blank = pureStr_list.count('')
    for i in range(blank) :
        pureStr_list.remove('')
    
    # 요구사항 4
    word_dict = {i : 0 for i in pureStr_list}
    for i in word_dict.keys() :
        for j in pureStr_list :
            if j == i :
                word_dict[i] += 1
        
    #print(pureStr.split())
    #print(pureStr_list)
    #return len(pureStr.split())
    return word_dict

# 요구사항 5
print(return_word_Num_dictionary('http://cse.koreatech.ac.kr'))
````
